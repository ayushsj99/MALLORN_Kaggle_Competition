# Making the Master Dataframe.

```python
import numpy as np
import pandas as pd

FILTERS = ["u", "g", "r", "i", "z", "y"]

EFF_WL = {
    "u": np.array([3641]),
    "g": np.array([4704]),
    "r": np.array([6155]),
    "i": np.array([7504]),
    "z": np.array([8695]),
    "y": np.array([10056])
}
```

```python
# 2) import necessary package and make necessary functions.

#Importing necessary package
!pip install extinction==0.4.7
from extinction import fitzpatrick99

#Defining function to de-extinct a set of flux values
def jurassic_park (flux, eff_wl, ebv):
    A_lambda = fitzpatrick99(eff_wl, ebv * 3.1) #3.1 = Standard Milky Way value
    flux_ext = flux * 10**((A_lambda)/2.5)
    return flux_ext, A_lambda
```

```python
import numpy as np
from scipy.signal import find_peaks

def build_extended_master_df(
    log_df,
    lc_prefix,        # "train_lc" or "test_lc"
    has_target=True
):
    master_rows_ext = []
    EPS = 1e-6

    for _, obj in log_df.iterrows():

        object_id = obj["object_id"]
        split = obj["split"]

        Z = obj["Z"]
        EBV = obj["EBV"]

        if has_target:
            target = obj["target"]

        lc_df = globals()[f"{lc_prefix}_{split}"]
        lc = lc_df[lc_df["object_id"] == object_id].copy()

        feat_row = {
            "object_id": object_id,
            "Z": Z,
            "EBV": EBV
        }

        if has_target:
            feat_row["target"] = target

        # -----------------------------
        # Containers for global physics
        # -----------------------------
        peak_times = {}
        decay_dists = []
        rise_mons = []
        total_secondary_peaks = 0

        # --------------------------------------------------
        # Per-band features
        # --------------------------------------------------
        for f in FILTERS:

            sub = lc[lc["Filter"] == f].sort_values("Time (MJD)")

            # -----------------------------
            # Defaults if insufficient data
            # -----------------------------
            if len(sub) < 5:
                for col in [
                    "n_obs","t_span","peak_time","rise_time","decay_time",
                    "flux_max","flux_mean","flux_std",
                    "mean_abs_diff","std_diff","max_jump",
                    "asymmetry","rise_frac","decay_frac","peak_frac",
                    "cv","peak_ratio","jump_ratio","diff_ratio","obs_density",
                    "decay_pl_slope","decay_pl_r2","decay_pl_dist",
                    "rise_monotonic_frac","rise_neg_frac",
                    "fluence_ratio","n_secondary_peaks"
                ]:
                    feat_row[f"{f}_{col}"] = 0
                continue

            time = sub["Time (MJD)"].values
            flux = sub["Flux"].values

            # De-extinction
            flux, _ = jurassic_park(flux, EFF_WL[f], EBV)

            peak_idx = np.argmax(flux)
            peak_time = time[peak_idx]
            peak_flux = flux[peak_idx]
            peak_times[f] = peak_time

            t_span = time[-1] - time[0]
            rise_time = peak_time - time[0]

            after_peak = flux[peak_idx:]
            if len(after_peak) > 1:
                half_flux = 0.5 * peak_flux
                decay_idx = np.where(after_peak <= half_flux)[0]
                decay_time = (
                    time[peak_idx + decay_idx[0]] - peak_time
                    if len(decay_idx) > 0 else 0
                )
            else:
                decay_time = 0

            flux_max = np.max(flux)
            flux_mean = np.mean(flux)
            flux_std = np.std(flux)

            diffs = np.abs(np.diff(flux))
            mean_abs_diff = np.mean(diffs)
            std_diff = np.std(diffs)
            max_jump = np.max(diffs)

            # -----------------------------
            # Physics: power-law decay
            # -----------------------------
            slope, r2 = 0.0, 0.0

            mask = (time > peak_time + 5)
            t_rel = time[mask] - peak_time
            f_rel = flux[mask]
            
            # ---- ROBUST GUARDS ----
            valid = (f_rel > 0) & np.isfinite(f_rel) & np.isfinite(t_rel)
            
            if valid.sum() >= 5:
                t_use = t_rel[valid]
                f_use = f_rel[valid]
            
                # require dynamic range (critical!)
                if np.std(np.log(f_use)) > 0.05:
                    try:
                        logt = np.log(t_use)
                        logf = np.log(f_use)
            
                        coef = np.polyfit(logt, logf, 1)
                        slope = coef[0]
            
                        pred = np.polyval(coef, logt)
                        ss_res = np.sum((logf - pred) ** 2)
                        ss_tot = np.sum((logf - logf.mean()) ** 2)
                        r2 = 1 - ss_res / (ss_tot + EPS)
            
                    except Exception:
                        slope, r2 = 0.0, 0.0
                        
            decay_dist = abs(slope + 5/3)
            decay_dists.append(decay_dist)

            # -----------------------------
            # Physics: rise monotonicity
            # -----------------------------
            rise_flux = flux[:peak_idx + 1]
            rise_diff = np.diff(rise_flux)
            rise_mon = np.mean(rise_diff > 0)
            rise_neg = np.mean(rise_diff < 0)
            rise_mons.append(rise_mon)

            # -----------------------------
            # Physics: fluence symmetry
            # -----------------------------
            flu_pre = np.trapz(flux[:peak_idx + 1], time[:peak_idx + 1])
            flu_post = np.trapz(flux[peak_idx:], time[peak_idx:])
            flu_ratio = flu_pre / (flu_post + EPS)

            # -----------------------------
            # Physics: secondary peaks
            # -----------------------------
            peaks, _ = find_peaks(flux, prominence=0.3 * peak_flux)
            n_secondary = max(0, len(peaks) - 1)
            total_secondary_peaks += n_secondary

            # -----------------------------
            # Store all features
            # -----------------------------
            feat_row.update({
                f"{f}_n_obs": len(flux),
                f"{f}_t_span": t_span,
                f"{f}_peak_time": peak_time,
                f"{f}_rise_time": rise_time,
                f"{f}_decay_time": decay_time,
                f"{f}_flux_max": flux_max,
                f"{f}_flux_mean": flux_mean,
                f"{f}_flux_std": flux_std,
                f"{f}_mean_abs_diff": mean_abs_diff,
                f"{f}_std_diff": std_diff,
                f"{f}_max_jump": max_jump,

                f"{f}_asymmetry": rise_time / (decay_time + EPS),
                f"{f}_rise_frac": rise_time / (t_span + EPS),
                f"{f}_decay_frac": decay_time / (t_span + EPS),
                f"{f}_peak_frac": (peak_time - time[0]) / (t_span + EPS),

                f"{f}_cv": flux_std / (flux_mean + EPS),
                f"{f}_peak_ratio": flux_max / (flux_mean + EPS),
                f"{f}_jump_ratio": max_jump / (flux_std + EPS),
                f"{f}_diff_ratio": mean_abs_diff / (flux_std + EPS),
                f"{f}_obs_density": len(flux) / (t_span + EPS),

                # physics
                f"{f}_decay_pl_slope": slope,
                f"{f}_decay_pl_r2": r2,
                f"{f}_decay_pl_dist": decay_dist,
                f"{f}_rise_monotonic_frac": rise_mon,
                f"{f}_rise_neg_frac": rise_neg,
                f"{f}_fluence_ratio": flu_ratio,
                f"{f}_n_secondary_peaks": n_secondary,
            })

        # --------------------------------------------------
        # Global physics features
        # --------------------------------------------------
        peak_list = [peak_times.get(b, 0) for b in FILTERS]
        band_order_violations = np.sum(np.diff(peak_list) < 0)

        feat_row["band_peak_order_violations"] = band_order_violations
        feat_row["mean_decay_pl_dist"] = np.mean(decay_dists) if decay_dists else 0
        feat_row["mean_rise_monotonic_frac"] = np.mean(rise_mons) if rise_mons else 0
        feat_row["total_secondary_peaks"] = total_secondary_peaks

        master_rows_ext.append(feat_row)

    df_ext = pd.DataFrame(master_rows_ext)
    df_ext = df_ext.replace([np.inf, -np.inf], np.nan).fillna(0)

    return df_ext
```

```python
import matplotlib.pyplot as plt

# pick one TDE and one non-TDE
tde_id = trainLog[trainLog["target"] == 1].iloc[0]["object_id"]
non_tde_id = trainLog[trainLog["target"] == 0].iloc[0]["object_id"]

def get_lc(object_id):
    split = trainLog.loc[trainLog["object_id"] == object_id, "split"].values[0]
    return globals()[f"train_lc_{split}"].query("object_id == @object_id")

lc_tde = get_lc(tde_id)
lc_non = get_lc(non_tde_id)
```

```python
FILTERS = ["u","g","r","i","z","y"]

def plot_lc(lc, title):
    plt.figure(figsize=(14,6))
    for f in FILTERS:
        sub = lc[lc["Filter"] == f]
        if len(sub) > 0:
            plt.scatter(
                sub["Time (MJD)"], sub["Flux"],
                s=12, label=f, alpha=0.7
            )
    plt.legend()
    plt.xlabel("Time (MJD)")
    plt.ylabel("Flux (μJy)")
    plt.title(title)
    plt.show()

plot_lc(lc_tde, "TDE Lightcurve (All Filters)")
plot_lc(lc_non, "Non-TDE Lightcurve (All Filters)")
```

```python
def plot_time_gaps(lc, title):
    plt.figure(figsize=(10,4))
    for f in FILTERS:
        sub = lc[lc["Filter"] == f].sort_values("Time (MJD)")
        if len(sub) > 1:
            gaps = np.diff(sub["Time (MJD)"])
            plt.hist(gaps, bins=30, alpha=0.5, label=f)
    plt.legend()
    plt.xlabel("Δ Time (days)")
    plt.title(title)
    plt.show()

plot_time_gaps(lc_tde, "TDE Observation Gaps")
```

```python
import pandas as pd
import numpy as np

FILTERS = ["u", "g", "r", "i", "z", "y"]

def analyze_lightcurve_lengths(
    log_df,
    lc_prefix="train_lc",
    max_objects=None
):
    rows = []

    iterator = log_df.iterrows()
    if max_objects:
        iterator = list(iterator)[:max_objects]

    for _, row in iterator:
        object_id = row["object_id"]
        split = row["split"]

        lc_df = globals()[f"{lc_prefix}_{split}"]
        lc = lc_df[lc_df["object_id"] == object_id]

        # ---------- global stats ----------
        all_times = lc["Time (MJD)"].values
        if len(all_times) > 1:
            global_n = len(all_times)
            global_span = all_times.max() - all_times.min()
        else:
            global_n = 0
            global_span = 0

        # ---------- per-filter stats ----------
        for f in FILTERS:
            sub = lc[lc["Filter"] == f]

            if len(sub) < 2:
                continue

            time = np.sort(sub["Time (MJD)"].values)

            gaps = np.diff(time)

            rows.append({
                "object_id": object_id,
                "filter": f,
                "n_obs": len(time),
                "t_span": time[-1] - time[0],
                "mean_gap": gaps.mean(),
                "max_gap": gaps.max(),
                "obs_density": len(time) / (time[-1] - time[0] + 1e-6),
                "global_n_obs": global_n,
                "global_t_span": global_span
            })

    return pd.DataFrame(rows)
```

# Adding more shape based features

```python
from scipy.stats import entropy
from scipy.signal import find_peaks

def compute_fwhm(time, flux, frac):
    """
    Width of peak at given fraction of peak flux
    """
    peak_flux = flux.max()
    thresh = frac * peak_flux
    mask = flux >= thresh

    if mask.sum() < 2:
        return 0.0

    return time[mask][-1] - time[mask][0]


def autocorr_lag(x, lag):
    if len(x) <= lag:
        return 0.0
    x1 = x[:-lag]
    x2 = x[lag:]
    if np.std(x1) < 1e-6 or np.std(x2) < 1e-6:
        return 0.0
    return np.corrcoef(x1, x2)[0, 1]


def structure_function_slope(time, flux):
    """
    Fit slope of structure function in log-log space
    """
    if len(flux) < 5:
        return 0.0

    dt = np.diff(time)
    df = np.diff(flux)

    valid = (dt > 0) & np.isfinite(df)
    if valid.sum() < 5:
        return 0.0

    sf = np.sqrt(np.mean(df[valid] ** 2))
    if sf <= 0:
        return 0.0

    return np.log(sf + 1e-6) / np.log(np.mean(dt[valid]) + 1e-6)
```

```python
def build_shape_only_features(
    log_df,
    lc_prefix
):
    rows = []
    EPS = 1e-6

    for _, obj in log_df.iterrows():
        object_id = obj["object_id"]
        split = obj["split"]
        EBV = obj["EBV"]

        lc_df = globals()[f"{lc_prefix}_{split}"]
        lc = lc_df[lc_df["object_id"] == object_id]

        row = {"object_id": object_id}

        for f in FILTERS:
            sub = lc[lc["Filter"] == f].sort_values("Time (MJD)")

            if len(sub) < 5:
                continue

            time = sub["Time (MJD)"].values
            flux = sub["Flux"].values
            flux, _ = jurassic_park(flux, EFF_WL[f], EBV)

            peak_idx = np.argmax(flux)
            peak_flux = flux[peak_idx]

            # ---- Shape features ----
            fwhm_50 = compute_fwhm(time, flux, 0.5)
            fwhm_25 = compute_fwhm(time, flux, 0.25)
            fwhm_ratio = fwhm_50 / (fwhm_25 + EPS)

            rise_time = time[peak_idx] - time[0]
            decay_time = time[-1] - time[peak_idx]
            t_span = time[-1] - time[0]

            peak_symmetry = abs(rise_time - decay_time) / (t_span + EPS)

            acf1 = autocorr_lag(flux, 1)
            acf2 = autocorr_lag(flux, 2)

            sf_slope = structure_function_slope(time, flux)

            peaks, props = find_peaks(flux, prominence=0.3 * peak_flux)
            if len(peaks) > 1:
                prom = props["prominences"]
                peak_entropy = entropy(prom / (prom.sum() + EPS))
            else:
                peak_entropy = 0.0

            row.update({
                f"{f}_fwhm_50": fwhm_50,
                f"{f}_fwhm_25": fwhm_25,
                f"{f}_fwhm_ratio": fwhm_ratio,
                f"{f}_peak_symmetry": peak_symmetry,
                f"{f}_acf_lag1": acf1,
                f"{f}_acf_lag2": acf2,
                f"{f}_sf_slope": sf_slope,
                f"{f}_peak_entropy": peak_entropy,
            })

        rows.append(row)

    df = pd.DataFrame(rows)
    return df.replace([np.inf, -np.inf], np.nan).fillna(0)
```

# Adding more shape related features

```python
import numpy as np

EPS = 1e-6

def compute_geometry_features(time, flux):
    feats = {
        "curvature": 0.0,
        "plateau_index": 0.0,
        "curve_asymmetry": 0.0,
        "arc_length": 0.0,
        "peak_shape_entropy": 0.0
    }

    # ---- Clean inputs ----
    mask = np.isfinite(time) & np.isfinite(flux)
    time = time[mask]
    flux = flux[mask]

    if len(flux) < 6 or np.ptp(flux) < EPS:
        return feats

    # sort
    idx = np.argsort(time)
    time = time[idx]
    flux = flux[idx]

    # -----------------------------
    # 1) Curvature (2nd derivative)
    # -----------------------------
    try:
        d1 = np.gradient(flux, time)
        d2 = np.gradient(d1, time)
        feats["curvature"] = np.nanmean(np.abs(d2)) / (np.ptp(flux) + EPS)
    except Exception:
        pass

    # peak
    p = np.argmax(flux)
    fmax = flux[p]

    # -----------------------------
    # 2) Plateau index
    # -----------------------------
    high = flux > 0.8 * fmax
    if high.sum() > 1:
        feats["plateau_index"] = (
            time[high][-1] - time[high][0]
        ) / (time[-1] - time[0] + EPS)

    # -----------------------------
    # 3) Curvature asymmetry
    # -----------------------------
    if 2 <= p <= len(flux) - 3:
        rise = np.nanmean(np.abs(d2[:p])) + EPS
        decay = np.nanmean(np.abs(d2[p:])) + EPS
        feats["curve_asymmetry"] = rise / decay

    # -----------------------------
    # 4) Arc length (path complexity)
    # -----------------------------
    dt = np.diff(time)
    df = np.diff(flux)
    arc = np.nansum(np.sqrt(dt**2 + df**2))
    feats["arc_length"] = arc / (time[-1] - time[0] + EPS)

    # -----------------------------
    # 5) Peak-centered shape entropy
    # -----------------------------
    w = max(3, int(0.3 * len(flux)))
    lo = max(0, p - w)
    hi = min(len(flux), p + w)

    slopes = np.diff(flux[lo:hi])
    slopes = slopes[np.isfinite(slopes)]

    if len(slopes) >= 6 and np.ptp(slopes) > EPS:
        hist, _ = np.histogram(slopes, bins=10, density=False)
        hist = hist.astype(float)
        hist /= (hist.sum() + EPS)
        feats["peak_shape_entropy"] = -np.sum(hist * np.log(hist + EPS))

    return feats
```

```python
def add_geometry_features_all_filters(df_ext, log_df, lc_prefix):
    df_ext = df_ext.copy()

    FILTERS = ["u", "g", "r", "i", "z", "y"]

    # initialize columns
    for f in FILTERS:
        for name in [
            "curvature",
            "plateau_index",
            "curve_asymmetry",
            "arc_length",
            "peak_shape_entropy"
        ]:
            df_ext[f"{f}_{name}"] = 0.0

    for _, row in log_df.iterrows():
        obj = row["object_id"]
        split = row["split"]
        ebv = row["EBV"]

        lc = globals()[f"{lc_prefix}_{split}"]

        for f in FILTERS:
            sub = lc[
                (lc["object_id"] == obj) &
                (lc["Filter"] == f)
            ].sort_values("Time (MJD)")

            if len(sub) < 6:
                continue

            time = sub["Time (MJD)"].values
            flux = sub["Flux"].values

            # de-extinction
            flux, _ = jurassic_park(flux, EFF_WL[f], ebv)

            feats = compute_geometry_features(time, flux)

            for k, v in feats.items():
                df_ext.loc[
                    df_ext.object_id == obj,
                    f"{f}_{k}"
                ] = v

    return df_ext
```

```python
def find_best_threshold(y_true, prob):
    thresholds = np.linspace(0.01, 0.99, 200)
    best_f1 = -1
    best_t = 0.5

    for t in thresholds:
        pred = (prob >= t).astype(int)
        report = classification_report(
            y_true, pred, output_dict=True, zero_division=0
        )
        f1 = report["1"]["f1-score"]
        if f1 > best_f1:
            best_f1 = f1
            best_t = t

    return best_t
```

# Adding more features to eliminate False Negatives

```python
import numpy as np

FILTERS = ["u", "g", "r", "i", "z", "y"]
EPS = 1e-6

def add_more_crossband_integrators(df):
    df = df.copy()

    # -------------------------------
    # Feature 3: Cross-band CV coherence
    # -------------------------------
    cv_cols = [f"{f}_cv" for f in FILTERS if f"{f}_cv" in df]
    cvs = df[cv_cols].values

    cv_mean = np.nanmean(cvs, axis=1)
    cv_std  = np.nanstd(cvs, axis=1)

    df["cv_crossband"] = cv_std / (cv_mean + EPS)

    # -------------------------------
    # Feature 4: Chromatic peak slope
    # -------------------------------
    peak_cols = [f"{f}_flux_max" for f in FILTERS if f"{f}_flux_max" in df]
    peaks = np.log1p(df[peak_cols].values)

    band_idx = np.arange(len(peak_cols))

    slopes = []
    for i in range(peaks.shape[0]):
        y = peaks[i]
        if np.all(np.isfinite(y)) and np.std(y) > 0:
            slope = np.polyfit(band_idx, y, 1)[0]
        else:
            slope = 0.0
        slopes.append(slope)

    df["chromatic_peak_slope"] = slopes

    # cleanup
    df = df.replace([np.inf, -np.inf], np.nan).fillna(0)
    return df
```

# Adding features to decrease the lift

```python
import numpy as np

EPS = 1e-6

def get_band_lightcurve(log_row, lc_prefix, band):
    """
    Returns time, de-extincted flux sorted by time.
    """
    obj = log_row["object_id"]
    split = log_row["split"]
    ebv = log_row["EBV"]

    lc_df = globals()[f"{lc_prefix}_{split}"]
    sub = lc_df[(lc_df["object_id"] == obj) & (lc_df["Filter"] == band)]

    if len(sub) < 5:
        return None, None

    sub = sub.sort_values("Time (MJD)")
    t = sub["Time (MJD)"].values
    f = sub["Flux"].values

    f, _ = jurassic_park(f, EFF_WL[band], ebv)

    return t, f
```

```python
def compute_pre_peak_dip_ratio(time, flux, window_frac=0.25):
    """
    Measures depth of dip before main peak.
    """
    peak_idx = np.argmax(flux)
    peak_flux = flux[peak_idx]

    if peak_idx < 3:
        return 0.0

    # look at last X% of rise
    lo = int((1 - window_frac) * peak_idx)
    pre_flux = flux[lo:peak_idx]

    if len(pre_flux) < 3:
        return 0.0

    dip = peak_flux - np.min(pre_flux)
    return dip / (peak_flux + EPS)
```

```python
def compute_global_peak_dominance(log_row, lc_prefix):
    """
    How dominant is the strongest peak compared to the entire LC background?
    """
    all_flux = []
    peak_fluxes = []

    for band in ["u","g","r","i","z","y"]:
        t, f = get_band_lightcurve(log_row, lc_prefix, band)
        if f is None:
            continue

        all_flux.extend(f.tolist())
        peak_fluxes.append(np.max(f))

    if len(all_flux) < 10 or len(peak_fluxes) == 0:
        return 0.0

    return np.max(peak_fluxes) / (np.median(all_flux) + EPS)
```

```python
def compute_peak_isolation(time, flux, frac=0.2):
    """
    Fraction of points outside peak window that are quiet.
    """
    peak_idx = np.argmax(flux)
    peak_flux = flux[peak_idx]

    if len(flux) < 8:
        return 0.0

    # approximate peak width via half max
    half_max = 0.5 * peak_flux
    above = np.where(flux >= half_max)[0]

    if len(above) < 2:
        return 0.0

    lo, hi = above[0], above[-1]

    outside = np.concatenate([flux[:lo], flux[hi+1:]])
    if len(outside) == 0:
        return 1.0

    quiet = np.mean(outside < frac * peak_flux)
    return quiet
```

```python
def add_peak_geometry_features(df_ext, log_df, lc_prefix):
    df_ext = df_ext.copy()

    df_ext["r_pre_peak_dip_ratio"] = 0.0
    df_ext["global_peak_dominance"] = 0.0
    df_ext["r_peak_isolation_score"] = 0.0

    for _, row in log_df.iterrows():
        obj = row["object_id"]
        idx = df_ext.index[df_ext.object_id == obj]
        if len(idx) == 0:
            continue
        idx = idx[0]

        # r-band features
        t, f = get_band_lightcurve(row, lc_prefix, "r")
        if f is not None:
            df_ext.loc[idx, "r_pre_peak_dip_ratio"] = compute_pre_peak_dip_ratio(t, f)
            df_ext.loc[idx, "r_peak_isolation_score"] = compute_peak_isolation(t, f)

        # global
        df_ext.loc[idx, "global_peak_dominance"] = compute_global_peak_dominance(
            row, lc_prefix
        )

    return df_ext
```

# More features to eliminate SN II and SN Ia

```python
import numpy as np

EPS = 1e-6

def add_crossband_physics_features(df):
    df = df.copy()

    # ============================
    # 1️⃣ Blue / Red Peak Ratio
    # ============================
    df["blue_red_peak_ratio"] = (
        df["u_flux_max"] + df["g_flux_max"]
    ) / (
        df["r_flux_max"] +
        df["i_flux_max"] +
        df["z_flux_max"] +
        df["y_flux_max"] +
        EPS
    )

    # ============================
    # 2️⃣ Red Persistence Index
    # ============================
    red_persistence_terms = []

    for b in ["r", "i", "z", "y"]:
        term = df[f"{b}_fluence_ratio"] / (1 + df[f"{b}_decay_frac"] + EPS)
        red_persistence_terms.append(term)

    df["red_persistence_index"] = np.mean(red_persistence_terms, axis=0)

    # ============================
    # 3️⃣ Blue Suppression Timing
    # ============================
    df["blue_suppression_timing"] = (
        df["u_rise_time"] + df["g_rise_time"]
    ) / (
        df["r_rise_time"] + df["i_rise_time"] + EPS
    )

    # ============================
    # Safety cleanup
    # ============================
    df.replace([np.inf, -np.inf], 0, inplace=True)
    df.fillna(0, inplace=True)

    return df
```

```python
def find_best_threshold(y_true, prob):
    thresholds = np.linspace(0.01, 0.99, 200)
    best_f1 = -1
    best_t = 0.5

    for t in thresholds:
        pred = (prob >= t).astype(int)
        report = classification_report(
            y_true, pred, output_dict=True, zero_division=0
        )
        f1 = report["1"]["f1-score"]
        if f1 > best_f1:
            best_f1 = f1
            best_t = t

    return best_t
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

oof_prob = np.zeros(len(y))
oof_pred = np.zeros(len(y))

fold_aucs = []
fold_thresholds = []
best_models = []

N_TRIALS = 20  # per fold (increase if time allows)

print(f"Samples: {len(y)}")
print(f"Positives: {y.sum()}  Negatives: {(y==0).sum()}")
print("Class ratio:", y.mean())

for fold, (tr_idx, val_idx) in enumerate(skf.split(X, y)):
    print(f"\n========== Fold {fold} ==========")

    X_tr, X_val = X[tr_idx], X[val_idx]
    y_tr, y_val = y[tr_idx], y[val_idx]

    best_auc = -np.inf
    best_model = None
    best_params = None

    for _ in range(N_TRIALS):
        params = {k: random.choice(v) for k, v in param_grid.items()}

        model = XGBClassifier(
            objective="binary:logistic",
            eval_metric="logloss",
            random_state=42,
            n_jobs=-1,
            scale_pos_weight=(y_tr == 0).sum() / (y_tr == 1).sum(),
            **params
        )

        model.fit(X_tr, y_tr)

        val_prob = model.predict_proba(X_val)[:, 1]
        auc = roc_auc_score(y_val, val_prob)

        if auc > best_auc:
            best_auc = auc
            best_model = model
            best_params = params

    # ---- threshold optimization (TRAIN ONLY) ----
    train_prob = best_model.predict_proba(X_tr)[:, 1]
    best_t = find_best_threshold(y_tr, train_prob)

    val_prob = best_model.predict_proba(X_val)[:, 1]
    val_pred = (val_prob >= best_t).astype(int)

    oof_prob[val_idx] = val_prob
    oof_pred[val_idx] = val_pred

    fold_aucs.append(best_auc)
    fold_thresholds.append(best_t)
    best_models.append(best_model)

    print("Best params:", best_params)
    print("Best threshold:", round(best_t, 3))
    print("Fold ROC-AUC:", round(best_auc, 4))
```

```python
def fp_lift_table(oof_df, trainLog):
    fp = oof_df[(oof_df.pred == 1) & (oof_df.target == 0)]
    
    fp_frac = fp["SpecType"].value_counts(normalize=True)
    base_frac = trainLog["SpecType"].value_counts(normalize=True)

    out = pd.concat([fp_frac, base_frac], axis=1)
    out.columns = ["FP_fraction", "Dataset_fraction"]
    out["FP_lift"] = out.FP_fraction / out.Dataset_fraction
    return out.sort_values("FP_lift", ascending=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

def average_lightcurve(
    df_group,          # subset of trainLog with object_id + split
    lc_prefix,
    filt,
    n_objects=80,
    n_grid=120
):
    curves = []
    t_min, t_max = np.inf, -np.inf

    sampled = df_group.sample(
        min(len(df_group), n_objects),
        random_state=42
    )

    for _, row in sampled.iterrows():
        lc_df = globals()[f"{lc_prefix}_{row['split']}"]
        ebv = row["EBV"]

        sub = lc_df[
            (lc_df["object_id"] == row["object_id"]) &
            (lc_df["Filter"] == filt)
        ]

        if len(sub) < 5:
            continue

        t = sub["Time (MJD)"].values
        f = sub["Flux"].values

        # de-extinction
        f, _ = jurassic_park(f, EFF_WL[filt], ebv)

        # align to peak
        peak_idx = np.argmax(f)
        t = t - t[peak_idx]

        t_min = min(t_min, t.min())
        t_max = max(t_max, t.max())

        curves.append((t, f))

    if len(curves) < 5:
        return None, None

    grid = np.linspace(t_min, t_max, n_grid)

    interp_flux = []
    for t, f in curves:
        interp = interp1d(
            t, f,
            bounds_error=False,
            fill_value=np.nan
        )(grid)
        interp_flux.append(interp)

    return grid, np.nanmean(interp_flux, axis=0)
```

```python
FILTERS = ["u", "g", "r", "i", "z", "y"]

def compare_spec_vs_tde(
    spectype,
    trainLog,
    lc_prefix="train_lc",
    n_objects=80
):
    print(f"\n🔍 Comparing TDE vs {spectype}")

    tde_df = trainLog[trainLog["target"] == 1]
    spec_df = trainLog[trainLog["SpecType"] == spectype]

    for f in FILTERS:
        t_tde, f_tde = average_lightcurve(
            tde_df, lc_prefix, f, n_objects
        )
        t_spec, f_spec = average_lightcurve(
            spec_df, lc_prefix, f, n_objects
        )

        if t_tde is None or t_spec is None:
            continue

        plt.figure(figsize=(7, 4))
        plt.plot(t_tde, f_tde, label="TDE", lw=2)
        plt.plot(t_spec, f_spec, label=spectype, lw=2, ls="--")

        plt.axvline(0, color="k", alpha=0.3)
        plt.title(f"{f}-band: TDE vs {spectype}")
        plt.xlabel("Time since peak (days)")
        plt.ylabel("Avg de-extincted flux")
        plt.legend()
        plt.tight_layout()
        plt.show()
```

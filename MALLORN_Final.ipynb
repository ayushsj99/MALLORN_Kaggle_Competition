# Making the Master Dataframe.

```python
import numpy as np
import pandas as pd

FILTERS = ["u", "g", "r", "i", "z", "y"]

EFF_WL = {
    "u": np.array([3641]),
    "g": np.array([4704]),
    "r": np.array([6155]),
    "i": np.array([7504]),
    "z": np.array([8695]),
    "y": np.array([10056])
}
```

```python
# 2) import necessary package and make necessary functions.

#Importing necessary package
!pip install extinction==0.4.7
from extinction import fitzpatrick99

#Defining function to de-extinct a set of flux values
def jurassic_park (flux, eff_wl, ebv):
    A_lambda = fitzpatrick99(eff_wl, ebv * 3.1) #3.1 = Standard Milky Way value
    flux_ext = flux * 10**((A_lambda)/2.5)
    return flux_ext, A_lambda
```

```python
import numpy as np
from scipy.signal import find_peaks

def build_extended_master_df(
    log_df,
    lc_prefix,        # "train_lc" or "test_lc"
    has_target=True
):
    master_rows_ext = []
    EPS = 1e-6

    for _, obj in log_df.iterrows():

        object_id = obj["object_id"]
        split = obj["split"]

        Z = obj["Z"]
        EBV = obj["EBV"]

        if has_target:
            target = obj["target"]

        lc_df = globals()[f"{lc_prefix}_{split}"]
        lc = lc_df[lc_df["object_id"] == object_id].copy()

        feat_row = {
            "object_id": object_id,
            "Z": Z,
            "EBV": EBV
        }

        if has_target:
            feat_row["target"] = target

        # -----------------------------
        # Containers for global physics
        # -----------------------------
        peak_times = {}
        decay_dists = []
        rise_mons = []
        total_secondary_peaks = 0

        # --------------------------------------------------
        # Per-band features
        # --------------------------------------------------
        for f in FILTERS:

            sub = lc[lc["Filter"] == f].sort_values("Time (MJD)")

            # -----------------------------
            # Defaults if insufficient data
            # -----------------------------
            if len(sub) < 5:
                for col in [
                    "n_obs","t_span","peak_time","rise_time","decay_time",
                    "flux_max","flux_mean","flux_std",
                    "mean_abs_diff","std_diff","max_jump",
                    "asymmetry","rise_frac","decay_frac","peak_frac",
                    "cv","peak_ratio","jump_ratio","diff_ratio","obs_density",
                    "decay_pl_slope","decay_pl_r2","decay_pl_dist",
                    "rise_monotonic_frac","rise_neg_frac",
                    "fluence_ratio","n_secondary_peaks"
                ]:
                    feat_row[f"{f}_{col}"] = 0
                continue

            time = sub["Time (MJD)"].values
            flux = sub["Flux"].values

            # De-extinction
            flux, _ = jurassic_park(flux, EFF_WL[f], EBV)

            peak_idx = np.argmax(flux)
            peak_time = time[peak_idx]
            peak_flux = flux[peak_idx]
            peak_times[f] = peak_time

            t_span = time[-1] - time[0]
            rise_time = peak_time - time[0]

            after_peak = flux[peak_idx:]
            if len(after_peak) > 1:
                half_flux = 0.5 * peak_flux
                decay_idx = np.where(after_peak <= half_flux)[0]
                decay_time = (
                    time[peak_idx + decay_idx[0]] - peak_time
                    if len(decay_idx) > 0 else 0
                )
            else:
                decay_time = 0

            flux_max = np.max(flux)
            flux_mean = np.mean(flux)
            flux_std = np.std(flux)

            diffs = np.abs(np.diff(flux))
            mean_abs_diff = np.mean(diffs)
            std_diff = np.std(diffs)
            max_jump = np.max(diffs)

            # -----------------------------
            # Physics: power-law decay
            # -----------------------------
            slope, r2 = 0.0, 0.0

            mask = (time > peak_time + 5)
            t_rel = time[mask] - peak_time
            f_rel = flux[mask]
            
            # ---- ROBUST GUARDS ----
            valid = (f_rel > 0) & np.isfinite(f_rel) & np.isfinite(t_rel)
            
            if valid.sum() >= 5:
                t_use = t_rel[valid]
                f_use = f_rel[valid]
            
                # require dynamic range (critical!)
                if np.std(np.log(f_use)) > 0.05:
                    try:
                        logt = np.log(t_use)
                        logf = np.log(f_use)
            
                        coef = np.polyfit(logt, logf, 1)
                        slope = coef[0]
            
                        pred = np.polyval(coef, logt)
                        ss_res = np.sum((logf - pred) ** 2)
                        ss_tot = np.sum((logf - logf.mean()) ** 2)
                        r2 = 1 - ss_res / (ss_tot + EPS)
            
                    except Exception:
                        slope, r2 = 0.0, 0.0
                        
            decay_dist = abs(slope + 5/3)
            decay_dists.append(decay_dist)

            # -----------------------------
            # Physics: rise monotonicity
            # -----------------------------
            rise_flux = flux[:peak_idx + 1]
            rise_diff = np.diff(rise_flux)
            rise_mon = np.mean(rise_diff > 0)
            rise_neg = np.mean(rise_diff < 0)
            rise_mons.append(rise_mon)

            # -----------------------------
            # Physics: fluence symmetry
            # -----------------------------
            flu_pre = np.trapz(flux[:peak_idx + 1], time[:peak_idx + 1])
            flu_post = np.trapz(flux[peak_idx:], time[peak_idx:])
            flu_ratio = flu_pre / (flu_post + EPS)

            # -----------------------------
            # Physics: secondary peaks
            # -----------------------------
            peaks, _ = find_peaks(flux, prominence=0.3 * peak_flux)
            n_secondary = max(0, len(peaks) - 1)
            total_secondary_peaks += n_secondary

            # -----------------------------
            # Store all features
            # -----------------------------
            feat_row.update({
                f"{f}_n_obs": len(flux),
                f"{f}_t_span": t_span,
                f"{f}_peak_time": peak_time,
                f"{f}_rise_time": rise_time,
                f"{f}_decay_time": decay_time,
                f"{f}_flux_max": flux_max,
                f"{f}_flux_mean": flux_mean,
                f"{f}_flux_std": flux_std,
                f"{f}_mean_abs_diff": mean_abs_diff,
                f"{f}_std_diff": std_diff,
                f"{f}_max_jump": max_jump,

                f"{f}_asymmetry": rise_time / (decay_time + EPS),
                f"{f}_rise_frac": rise_time / (t_span + EPS),
                f"{f}_decay_frac": decay_time / (t_span + EPS),
                f"{f}_peak_frac": (peak_time - time[0]) / (t_span + EPS),

                f"{f}_cv": flux_std / (flux_mean + EPS),
                f"{f}_peak_ratio": flux_max / (flux_mean + EPS),
                f"{f}_jump_ratio": max_jump / (flux_std + EPS),
                f"{f}_diff_ratio": mean_abs_diff / (flux_std + EPS),
                f"{f}_obs_density": len(flux) / (t_span + EPS),

                # physics
                f"{f}_decay_pl_slope": slope,
                f"{f}_decay_pl_r2": r2,
                f"{f}_decay_pl_dist": decay_dist,
                f"{f}_rise_monotonic_frac": rise_mon,
                f"{f}_rise_neg_frac": rise_neg,
                f"{f}_fluence_ratio": flu_ratio,
                f"{f}_n_secondary_peaks": n_secondary,
            })

        # --------------------------------------------------
        # Global physics features
        # --------------------------------------------------
        peak_list = [peak_times.get(b, 0) for b in FILTERS]
        band_order_violations = np.sum(np.diff(peak_list) < 0)

        feat_row["band_peak_order_violations"] = band_order_violations
        feat_row["mean_decay_pl_dist"] = np.mean(decay_dists) if decay_dists else 0
        feat_row["mean_rise_monotonic_frac"] = np.mean(rise_mons) if rise_mons else 0
        feat_row["total_secondary_peaks"] = total_secondary_peaks

        master_rows_ext.append(feat_row)

    df_ext = pd.DataFrame(master_rows_ext)
    df_ext = df_ext.replace([np.inf, -np.inf], np.nan).fillna(0)

    return df_ext
```

```python
import matplotlib.pyplot as plt

# pick one TDE and one non-TDE
tde_id = trainLog[trainLog["target"] == 1].iloc[0]["object_id"]
non_tde_id = trainLog[trainLog["target"] == 0].iloc[0]["object_id"]

def get_lc(object_id):
    split = trainLog.loc[trainLog["object_id"] == object_id, "split"].values[0]
    return globals()[f"train_lc_{split}"].query("object_id == @object_id")

lc_tde = get_lc(tde_id)
lc_non = get_lc(non_tde_id)
```

```python
FILTERS = ["u","g","r","i","z","y"]

def plot_lc(lc, title):
    plt.figure(figsize=(14,6))
    for f in FILTERS:
        sub = lc[lc["Filter"] == f]
        if len(sub) > 0:
            plt.scatter(
                sub["Time (MJD)"], sub["Flux"],
                s=12, label=f, alpha=0.7
            )
    plt.legend()
    plt.xlabel("Time (MJD)")
    plt.ylabel("Flux (μJy)")
    plt.title(title)
    plt.show()

plot_lc(lc_tde, "TDE Lightcurve (All Filters)")
plot_lc(lc_non, "Non-TDE Lightcurve (All Filters)")
```

```python
def plot_time_gaps(lc, title):
    plt.figure(figsize=(10,4))
    for f in FILTERS:
        sub = lc[lc["Filter"] == f].sort_values("Time (MJD)")
        if len(sub) > 1:
            gaps = np.diff(sub["Time (MJD)"])
            plt.hist(gaps, bins=30, alpha=0.5, label=f)
    plt.legend()
    plt.xlabel("Δ Time (days)")
    plt.title(title)
    plt.show()

plot_time_gaps(lc_tde, "TDE Observation Gaps")
```

```python
import pandas as pd
import numpy as np

FILTERS = ["u", "g", "r", "i", "z", "y"]

def analyze_lightcurve_lengths(
    log_df,
    lc_prefix="train_lc",
    max_objects=None
):
    rows = []

    iterator = log_df.iterrows()
    if max_objects:
        iterator = list(iterator)[:max_objects]

    for _, row in iterator:
        object_id = row["object_id"]
        split = row["split"]

        lc_df = globals()[f"{lc_prefix}_{split}"]
        lc = lc_df[lc_df["object_id"] == object_id]

        # ---------- global stats ----------
        all_times = lc["Time (MJD)"].values
        if len(all_times) > 1:
            global_n = len(all_times)
            global_span = all_times.max() - all_times.min()
        else:
            global_n = 0
            global_span = 0

        # ---------- per-filter stats ----------
        for f in FILTERS:
            sub = lc[lc["Filter"] == f]

            if len(sub) < 2:
                continue

            time = np.sort(sub["Time (MJD)"].values)

            gaps = np.diff(time)

            rows.append({
                "object_id": object_id,
                "filter": f,
                "n_obs": len(time),
                "t_span": time[-1] - time[0],
                "mean_gap": gaps.mean(),
                "max_gap": gaps.max(),
                "obs_density": len(time) / (time[-1] - time[0] + 1e-6),
                "global_n_obs": global_n,
                "global_t_span": global_span
            })

    return pd.DataFrame(rows)
```

# Training an XGBModel

```python
import numpy as np
import pandas as pd

import shap
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score, classification_report
```

```python
X_train = train_df_ext.drop(columns=["object_id", "target"])
y_train = train_df_ext["target"].values

X_test = test_df_ext.drop(columns=["object_id"])

feature_names = X_train.columns.tolist()

X_train = X_train.values
X_test  = X_test.values
```

```python
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

model = XGBClassifier(
    n_estimators=400,
    max_depth=4,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    min_child_weight=5,
    gamma=0.1,
    objective="binary:logistic",
    eval_metric="logloss",
    scale_pos_weight=scale_pos_weight,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)
```

```python
train_prob = model.predict_proba(X_train)[:, 1]
print("Train ROC-AUC:", roc_auc_score(y_train, train_prob))
```

```python
def model_predict(X):
    return model.predict_proba(X)[:, 1]
```

```python
explainer = shap.Explainer(
    model_predict,
    X_train,
    feature_names=feature_names
)

shap_values = explainer(X_train)
```

```python
shap_importance = np.abs(shap_values.values).mean(axis=0)

shap_df = (
    pd.DataFrame({
        "feature": feature_names,
        "mean_abs_shap": shap_importance
    })
    .sort_values("mean_abs_shap", ascending=False)
)

shap_df.head(20)
```

```python
TOP_K = 80
selected_features = shap_df["feature"].iloc[:TOP_K].tolist()

print("Selected features:", selected_features)
```

```python
X_train_red = train_df_ext[selected_features].values
X_test_red  = test_df_ext[selected_features].values
y = train_df_ext["target"].values
```

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import StratifiedKFold, RandomizedSearchCV
from sklearn.metrics import (
    classification_report,
    roc_auc_score,
    f1_score
)

from xgboost import XGBClassifier
```

```python
param_dist = {
    "max_depth": [3, 4, 5],
    "learning_rate": [0.01, 0.03, 0.05],
    "n_estimators": [300, 500, 700],
    "subsample": [0.7, 0.8, 0.9],
    "colsample_bytree": [0.6, 0.8, 1.0],
    "min_child_weight": [1, 5, 10],
    "gamma": [0, 0.1, 0.3]
}
```

```python
N_SPLITS = 5
skf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=42)

oof_prob = np.zeros(len(y))
oof_pred = np.zeros(len(y))

fold_aucs = []
fold_thresholds = []
best_models = []
```

```python
for fold, (tr_idx, va_idx) in enumerate(skf.split(X_train_red, y)):
    print(f"\n========== Fold {fold} ==========")

    X_tr, X_va = X_train_red[tr_idx], X_train_red[va_idx]
    y_tr, y_va = y[tr_idx], y[va_idx]

    scale_pos_weight = (y_tr == 0).sum() / (y_tr == 1).sum()

    base_model = XGBClassifier(
        objective="binary:logistic",
        eval_metric="logloss",
        scale_pos_weight=scale_pos_weight,
        random_state=42,
        n_jobs=-1
    )

    # ---- hyperparameter tuning (TRAIN ONLY) ----
    search = RandomizedSearchCV(
        estimator=base_model,
        param_distributions=param_dist,
        n_iter=20,
        scoring="roc_auc",
        cv=3,
        verbose=0,
        n_jobs=-1,
        random_state=42
    )

    search.fit(X_tr, y_tr)
    model = search.best_estimator_

    print("Best params:", search.best_params_)

    # ---- TRAIN probabilities (for threshold tuning) ----
    prob_tr = model.predict_proba(X_tr)[:, 1]

    # ---- threshold optimization (TRAIN ONLY) ----
    thresholds = np.linspace(0.05, 0.6, 150)
    f1s = [f1_score(y_tr, (prob_tr >= t).astype(int)) for t in thresholds]
    best_t = thresholds[np.argmax(f1s)]

    fold_thresholds.append(best_t)
    print("Best threshold:", round(best_t, 3))

    # ---- VALIDATION ----
    prob_va = model.predict_proba(X_va)[:, 1]
    pred_va = (prob_va >= best_t).astype(int)

    oof_prob[va_idx] = prob_va
    oof_pred[va_idx] = pred_va

    auc = roc_auc_score(y_va, prob_va)
    fold_aucs.append(auc)

    print("Fold ROC-AUC:", round(auc, 4))

    best_models.append(model)
```

```python
print("\n==============================")
print("Mean CV ROC-AUC:", np.mean(fold_aucs))
print("Std  CV ROC-AUC:", np.std(fold_aucs))
print("Mean optimal threshold:", np.mean(fold_thresholds))
print("==============================\n")

print("OOF Classification Report:\n")
print(classification_report(y, oof_pred, digits=4))

print("OOF ROC-AUC:", roc_auc_score(y, oof_prob))
```

```python
FINAL_THRESHOLD = float(np.mean(fold_thresholds))

final_model = XGBClassifier(
    **best_models[np.argmax(fold_aucs)].get_params()
)

final_model.fit(X_train_red, y)
```

```python
test_prob = final_model.predict_proba(X_test_red)[:, 1]
test_pred = (test_prob >= FINAL_THRESHOLD).astype(int)

submission = pd.DataFrame({
    "object_id": test_df_ext["object_id"],
    "target": test_pred
})

submission.to_csv("submission.csv", index=False)
submission.head()
```

```python
from scipy.stats import entropy
from scipy.signal import find_peaks

def compute_fwhm(time, flux, frac):
    """
    Width of peak at given fraction of peak flux
    """
    peak_flux = flux.max()
    thresh = frac * peak_flux
    mask = flux >= thresh

    if mask.sum() < 2:
        return 0.0

    return time[mask][-1] - time[mask][0]


def autocorr_lag(x, lag):
    if len(x) <= lag:
        return 0.0
    x1 = x[:-lag]
    x2 = x[lag:]
    if np.std(x1) < 1e-6 or np.std(x2) < 1e-6:
        return 0.0
    return np.corrcoef(x1, x2)[0, 1]


def structure_function_slope(time, flux):
    """
    Fit slope of structure function in log-log space
    """
    if len(flux) < 5:
        return 0.0

    dt = np.diff(time)
    df = np.diff(flux)

    valid = (dt > 0) & np.isfinite(df)
    if valid.sum() < 5:
        return 0.0

    sf = np.sqrt(np.mean(df[valid] ** 2))
    if sf <= 0:
        return 0.0

    return np.log(sf + 1e-6) / np.log(np.mean(dt[valid]) + 1e-6)
```

```python
def build_shape_only_features(
    log_df,
    lc_prefix
):
    rows = []
    EPS = 1e-6

    for _, obj in log_df.iterrows():
        object_id = obj["object_id"]
        split = obj["split"]
        EBV = obj["EBV"]

        lc_df = globals()[f"{lc_prefix}_{split}"]
        lc = lc_df[lc_df["object_id"] == object_id]

        row = {"object_id": object_id}

        for f in FILTERS:
            sub = lc[lc["Filter"] == f].sort_values("Time (MJD)")

            if len(sub) < 5:
                continue

            time = sub["Time (MJD)"].values
            flux = sub["Flux"].values
            flux, _ = jurassic_park(flux, EFF_WL[f], EBV)

            peak_idx = np.argmax(flux)
            peak_flux = flux[peak_idx]

            # ---- Shape features ----
            fwhm_50 = compute_fwhm(time, flux, 0.5)
            fwhm_25 = compute_fwhm(time, flux, 0.25)
            fwhm_ratio = fwhm_50 / (fwhm_25 + EPS)

            rise_time = time[peak_idx] - time[0]
            decay_time = time[-1] - time[peak_idx]
            t_span = time[-1] - time[0]

            peak_symmetry = abs(rise_time - decay_time) / (t_span + EPS)

            acf1 = autocorr_lag(flux, 1)
            acf2 = autocorr_lag(flux, 2)

            sf_slope = structure_function_slope(time, flux)

            peaks, props = find_peaks(flux, prominence=0.3 * peak_flux)
            if len(peaks) > 1:
                prom = props["prominences"]
                peak_entropy = entropy(prom / (prom.sum() + EPS))
            else:
                peak_entropy = 0.0

            row.update({
                f"{f}_fwhm_50": fwhm_50,
                f"{f}_fwhm_25": fwhm_25,
                f"{f}_fwhm_ratio": fwhm_ratio,
                f"{f}_peak_symmetry": peak_symmetry,
                f"{f}_acf_lag1": acf1,
                f"{f}_acf_lag2": acf2,
                f"{f}_sf_slope": sf_slope,
                f"{f}_peak_entropy": peak_entropy,
            })

        rows.append(row)

    df = pd.DataFrame(rows)
    return df.replace([np.inf, -np.inf], np.nan).fillna(0)
```

# Adding more shape related features

```python
import numpy as np

EPS = 1e-6

def compute_geometry_features(time, flux):
    feats = {
        "curvature": 0.0,
        "plateau_index": 0.0,
        "curve_asymmetry": 0.0,
        "arc_length": 0.0,
        "peak_shape_entropy": 0.0
    }

    # ---- Clean inputs ----
    mask = np.isfinite(time) & np.isfinite(flux)
    time = time[mask]
    flux = flux[mask]

    if len(flux) < 6 or np.ptp(flux) < EPS:
        return feats

    # sort
    idx = np.argsort(time)
    time = time[idx]
    flux = flux[idx]

    # -----------------------------
    # 1) Curvature (2nd derivative)
    # -----------------------------
    try:
        d1 = np.gradient(flux, time)
        d2 = np.gradient(d1, time)
        feats["curvature"] = np.nanmean(np.abs(d2)) / (np.ptp(flux) + EPS)
    except Exception:
        pass

    # peak
    p = np.argmax(flux)
    fmax = flux[p]

    # -----------------------------
    # 2) Plateau index
    # -----------------------------
    high = flux > 0.8 * fmax
    if high.sum() > 1:
        feats["plateau_index"] = (
            time[high][-1] - time[high][0]
        ) / (time[-1] - time[0] + EPS)

    # -----------------------------
    # 3) Curvature asymmetry
    # -----------------------------
    if 2 <= p <= len(flux) - 3:
        rise = np.nanmean(np.abs(d2[:p])) + EPS
        decay = np.nanmean(np.abs(d2[p:])) + EPS
        feats["curve_asymmetry"] = rise / decay

    # -----------------------------
    # 4) Arc length (path complexity)
    # -----------------------------
    dt = np.diff(time)
    df = np.diff(flux)
    arc = np.nansum(np.sqrt(dt**2 + df**2))
    feats["arc_length"] = arc / (time[-1] - time[0] + EPS)

    # -----------------------------
    # 5) Peak-centered shape entropy
    # -----------------------------
    w = max(3, int(0.3 * len(flux)))
    lo = max(0, p - w)
    hi = min(len(flux), p + w)

    slopes = np.diff(flux[lo:hi])
    slopes = slopes[np.isfinite(slopes)]

    if len(slopes) >= 6 and np.ptp(slopes) > EPS:
        hist, _ = np.histogram(slopes, bins=10, density=False)
        hist = hist.astype(float)
        hist /= (hist.sum() + EPS)
        feats["peak_shape_entropy"] = -np.sum(hist * np.log(hist + EPS))

    return feats
```

```python
def add_geometry_features_all_filters(df_ext, log_df, lc_prefix):
    df_ext = df_ext.copy()

    FILTERS = ["u", "g", "r", "i", "z", "y"]

    # initialize columns
    for f in FILTERS:
        for name in [
            "curvature",
            "plateau_index",
            "curve_asymmetry",
            "arc_length",
            "peak_shape_entropy"
        ]:
            df_ext[f"{f}_{name}"] = 0.0

    for _, row in log_df.iterrows():
        obj = row["object_id"]
        split = row["split"]
        ebv = row["EBV"]

        lc = globals()[f"{lc_prefix}_{split}"]

        for f in FILTERS:
            sub = lc[
                (lc["object_id"] == obj) &
                (lc["Filter"] == f)
            ].sort_values("Time (MJD)")

            if len(sub) < 6:
                continue

            time = sub["Time (MJD)"].values
            flux = sub["Flux"].values

            # de-extinction
            flux, _ = jurassic_park(flux, EFF_WL[f], ebv)

            feats = compute_geometry_features(time, flux)

            for k, v in feats.items():
                df_ext.loc[
                    df_ext.object_id == obj,
                    f"{f}_{k}"
                ] = v

    return df_ext
```
